```python
# Q 3.2

import numpy as np
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
planets = {}

for i in range(len(masses)):
    M = masses[i]
    r = radii[i]
    v = np.sqrt(2 * G * M / (r * 1000))
    print(f"The escape velocity of {names[i]} is {v:.2f} m/s.")
    planets[names[i]] = v

print(planets)
```

```python
# Q 3.7
import numpy as np

def angs(d,D):
    d = float(input('Size of object:'))
    D = float(input('Distance to object:'))
    delta = 2 * np.arctan(d / (2 * D))
    return delta

angs(d,D)
print(f'The angular size of the object is {delta} arcsec.')
```

```python
# Q 3.10
from math import pi

r = float(input('Radius:'))
h = float(input('Height:'))

def cylinder(r, h):
    A = 2 * pi * r**2 + 2 * pi * r * h
    V = pi * r**2 * h
    s = f'A = {A}, V = {V}'
    return s

print(f'The area and volume are: {cylinder(r,h)}')
```

```python
# Q 3.17 (pending)

import numpy as np

a = np.random.normal(0,1,1000)

def stdev(a):
    N = len(a)
    sig = np.sqrt(1/N * sum((a - (1/N * sum(a)))**2))
    return sig

print(f'Manual: {stdev(a)}')
print(f'With NumPy {np.std(a)}')
```
