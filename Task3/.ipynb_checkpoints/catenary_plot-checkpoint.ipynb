```python
import matplotlib.pyplot as plt
import numpy as np

c = []
for x in range(-5,6):
    c.append(np.cosh(x))

plt.scatter(range(-5,6), c, marker='+', s=80, label='x-points')
plt.plot(range(-5,6), c, color='grey', label='x-approx')
plt.title('$cosh(x)$')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()

C = np.array(c)

plt.scatter(range(-5,6), C, marker='+', s=80, label='x-points')
plt.plot(range(-5,6), C, color='grey', label='x-approx')
plt.title('$cosh(x)$')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```
