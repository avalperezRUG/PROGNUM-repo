```python
# Q 5.6
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

ab = np.concatenate((a,b))
print(ab)
print()

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print(np.concatenate((a, b)))     # axis=0 by default
print()

print(np.concatenate((a, b), axis=0))     # concatenates it below
print()

print(np.concatenate((a, b), axis=None))     # concatenates it to the right
print()
```

```python
# Q 5.9
from imageio.v2 import imwrite, imread
import matplotlib.pyplot as plt
import numpy as np

img = imread('m101BW.jpg')
fig = plt.figure(figsize=(7,7))
frame = fig.add_subplot(1,1,1)
frame.imshow(img, cmap='jet')
frame.contour(img, levels=np.percentile(img, [85, 90, 99]), colors=['y', 'b', 'r'])
fig.show()
```

```python
# Q 5.9 Part 2
from imageio.v2 import imwrite, imread
import matplotlib.pyplot as plt
import numpy as np

img = imread('m101BW.jpg')

s = np.percentile(img, 90)
imgnew = np.where(img > s, img, np.nan)
plt.imshow(imgnew)
```

```python
# Q 5.10
import numpy as np
import matplotlib.pyplot as plt

M = np.loadtxt('ngc6946.txt')
M.shape = (40, 50, 70)


zmid = M[:, 25, 35]
print(zmid, '\n max =', zmid.max(), ', min =', zmid.min(), ', sum =', zmid.sum(), ', mean =', zmid.mean(), ', stdev =', zmid.std())

plt.plot(zmid)
plt.text(11, .059,
         f'Max = {zmid.max():f}')
plt.text(13, .006,
         f'Mean = {zmid.mean():f}')
plt.show()
```

```python
# Q 5.11
import numpy as np
import matplotlib.pyplot as plt

M = np.loadtxt('ngc6946.txt')
M.shape = (40, 50, 70)

plane = M[0]
m = plane.mean()
std = plane.std()

plt.imshow(plane, origin='lower')
plt.colorbar()
plt.contour(plane, levels=[m, m + std, m + 2*std], colors=['y', 'b', 'r'])
plt.show()
```
