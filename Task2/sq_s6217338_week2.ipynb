```python
from math import atan, atan2, degrees
x = float(input("x-value:"))
y = float(input("y-value:"))

angle = degrees(atan2(y, x))

print(f"The angle this point, connected to the origin, has with the positive x-axis is {round(angle, 2)}°.")

s = f"This point belongs to "
if 0 <= angle <= 90:
    print(s+"QI.")
elif 90 < angle <= 180:
    print(s+"QII.")
elif -180 < angle <= -90:
    print(s+"QIII.")
else:
    print(s+"QIV.")
```
