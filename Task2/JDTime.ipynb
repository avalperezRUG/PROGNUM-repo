```python
print("Please enter the initial date:")
Y1 = float(input("Year:"))
M1 = float(input("Month:"))
D1 = float(input("Day:"))

JD1 = 367*Y1 -7*(Y1+(M1+9)//12)//4 - 3*((Y1+(M1-9)//7)//100 + 1)//4 + (275*M1)//9 + D1 + 1721029-0.5

print("Now enter the final date:")
Y2 = float(input("Year:"))
M2 = float(input("Month:"))
D2 = float(input("Day:"))

JD2 = 367*Y2 -7*(Y2+(M2+9)//12)//4 - 3*((Y2+(M2-9)//7)//100 + 1)//4 + (275*M2)//9 + D2 + 1721029-0.5

time = JD2 - JD1
print(f"The time, in days, elapsed between these to dates is {time}")
```
